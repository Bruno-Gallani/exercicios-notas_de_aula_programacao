##Exercício 1 - Área do círculo
.......png

```python
import math  #comando usado para fazer a maior parte das operações matemáticas

raio = float(input('raio = \t'))  #raio: é uma variável    #float: comando para tornar uma variável um número real

#input: comando usado para aparecer uma coisa no monitor, pedindo a digitação de algo (recepção de dados do usuário)  #\t: efetua o espaçamento durante o input/print

área = ((math.pi) * (raio**2))  #área: é uma variável

print(f"área = \t{área:.4f}")  #print: comando usado para aparecer o resultado final no monitor (saída de dados)

# f"________{____:.Xf}" ---> comando usado para formatar uma determinada variável, de acordo com as preferências do usuário    #f: format

# :.Xf ----> se escreve isso para definir o número de casas decimais que um determinado valor deve possuir
```

##Exercício 2 - Paralelepípedos
image.png

```python
import math

a = float(input())

b = float(input())

c = float(input())

V = Volume = (a*b*c)

d = Diagonal = math.sqrt(a**2 + b**2 + c**2)

#  ** = símbolo de potenciação
#ex: x**2 = x²

#math.sqrt ---> comando usado para efetuar o cálculo de raíz quadrada
#exemplos: math.sqrt(x) = √x          math.sqrt(x**2) = √x² = x

print(f"Volume: \t{V:.2f}")

print(f"Diagonal: \t{d:.2f}")
```

##Exercício 3 - Conversão de temperaturas
image.png

```python
import math

C = float(input())

#C = Celsius

Fahrenheit = int(math.floor(1.8*C+32))

Kelvin = int(math.floor(C+273.15))

print(f"{Fahrenheit:.0f}", "F")

print(f"{Kelvin:.0f}", "K")
```

##Exercício 4 - Exponencial, raíz e fração
image.png

```python
import math

a = int(input())

#int: comando para tornar uma variável como um número inteiro

b = int(input())

x = float(input())

y = float(input())

expressão = a + b**x - math.sqrt(b) + (a+b) / (x-y)

# a + b^x - √b + (a+b/x-y)  ------> fração

#para utilizar frações no colab, não é necessário nenhum comando, é só utilizar os parêntesis no denominador (de cima) e no numerador (de baixo)
# e também utilizar a barra, para a divisão dos dois termos

print(f"expressão: {expressão:.2f}")

#  print("____________") ----> forma certa de se escrever
```

##Exercício 7 - O maior número inteiro

```python
import math

x = float(input())

y = float(input())

z = float(input())

O_maior_inteiro_entre_xy = ((x+y) / 2) + (math.fabs(y-x) / 2)

O_maior_inteiro_entre_xyz = ((O_maior_inteiro_entre_xy+z) / 2) + (math.fabs(z-O_maior_inteiro_entre_xy) / 2)

print("O maior inteiro:", int(O_maior_inteiro_entre_xyz))
```

##Exercício 8 - Equação do 2º grau

```python
import math

a = float(input())

a != 0

b = float (input())

c = float (input())

x = float (input())

delta = b**2 - 4*a*c

delta >= 0

primeiraRaiz = (-b + math.sqrt(delta)) / (2*a)

segundaRaiz = (-b -math.sqrt(delta)) / (2*a)

equação2ºgrau = a*x**2 + b*x + c

print ("raizes:", f"{primeiraRaiz:.2f}", f"{segundaRaiz:.2f}")

print ("resultado para x =", f"{x:.2f}" + ":", f"{equação2ºgrau:.2f}")
```

##Exercício 9 - Encriptação

```python
N1 = int(input())
#N1 = número não encriptado

alg_unidade_milhar_N2 = N1 // 1000
#N2 = número encriptado
#alg = algarismo

if (0 <= alg_unidade_milhar_N2 < 9):
  alg_unidade_milhar_N2 = alg_unidade_milhar_N2 + 1

else:
  alg_unidade_milhar_N2 = 0

ordem_de_centena_N1 = N1 % 1000
alg_centena_N2 = ordem_de_centena_N1 // 100
#ordem de centena: 1200 ----> 200

if (0 <= alg_centena_N2 < 9):
  alg_centena_N2 = alg_centena_N2 + 1
else:
  alg_centena_N2 = 0

ordem_de_dezena_N1 = N1 % 100
alg_dezena_N2 = ordem_de_dezena_N1 // 10

if (0 <= alg_dezena_N2 < 9):
  alg_dezena_N2 = alg_dezena_N2 + 1
else:
  alg_dezena_N2 = 0

ordem_da_unidade_N1 = N1 % 10
alg_unidade_N2 = ordem_da_unidade_N1

if (0 <= alg_unidade_N2 < 9):
  alg_unidade_N2 = alg_unidade_N2 + 1
else:
  alg_unidade_N2 = 0

N2 = str(alg_unidade_milhar_N2) + str(alg_centena_N2) + str(alg_dezena_N2) + str(alg_unidade_N2)

print(N2)
```
