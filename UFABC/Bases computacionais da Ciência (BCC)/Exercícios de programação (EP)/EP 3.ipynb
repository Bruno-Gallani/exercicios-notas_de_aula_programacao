#Exercício 4 - Jogo do PIM
image.png

```python
import numpy as np

N = int(input())
while (N < 0):
  print("O valor a ser digitado deve ser um número inteiro maior ou igual a zero")
  N = int(input())

numeros_naturais = np.arange(1, (4*N)+1)       #todos os números naturais no intervalo de 1 a 4N
                                               #[1, 2, 3, .........., 4*N]


MultiplosDe4 = np.arange(4, (4*N)+1, 4)        #[4, 8, ..................., 4*N]


numeros_naturais = np.setdiff1d(numeros_naturais, MultiplosDe4)           #atualização do valor da variável numeros_naturais
                                                                          #np.setdiff1d = função para executar a diferença entre dois conjuntos
                                                                          #[numeros_naturais] - [MultiplosDe4]


numeros_naturais = np.array_split(numeros_naturais, N)          #np.array_split: divide uma lista, criada pela função np.arange (ou np.array), em sublistas
                                                                #sintaxe: np.array_split(lista que se pretende dividir, número de sublistas)

for i in range(0, N):
  print(*numeros_naturais[i], "PIM")        #o asterisco, quando usado em listas, remove os colchetes
```

#Exercício 6 - É Fatorial?
image.png

```python
import math

import numpy as np

N = int(input())
while (N < 0):
  print("O valor a ser digitado deve ser um número inteiro maior ou igual a zero")
  N = int(input())


X = np.arange(0, 1000)      #X ---> todos os números compreendidos no intervalo de 0 a 1000 [0, 1, ....., 1000]

if any(math.factorial(i) == N for i in X):       #se o fatorial de qualquer número de 0 a 1000 for igual à entrada (N), ou seja, se N = X!
  print("Verdadeiro")                            #printar verdadeiro
                                                 #sintaxe da função any: any(elemento da lista, em função de i, for i in lista)
                                                 #https://pythonhelp.wordpress.com/2012/11/06/all-e-any/
else:
  print("Falso")
```

#Exercício 8 - Proporção de adultos e médias das idades
image.png

```python
N = int(input())
while not (1 <= N <= 100):           #enquanto N não estiver compreendido entre 1 e 100, inclusive
  print("A quantidade de pessoas deve ser entre 1 e 100, inclusive")
  N = int(input())                   #N: representa a quantidade de pessoas

soma_idades = 0

numero_adultos = 0

for i in range(0, N):
  idade = int(input())
  0 <= idade <= 130
  soma_idades = soma_idades + idade
  if (idade >= 18):
    numero_adultos = numero_adultos + 1

media = soma_idades / N

proporcao = (numero_adultos / N) * 100

print("%.1f"%proporcao + "%")
print("%.1f"%media)
```

#Exercício 9 - Falha do motor
image.png

```python
#Forma 1 - Feita pelo professor

def falha_motor(N):
  atual = int(input())
  for i in range(2,N+1):
    prox = int(input())
    if (prox < atual):
      return i
    atual = prox
  return 0

N = int (input())
print(falha_motor(N))
```

```python
#Forma 1

P = int(input())        #altura do pulo do sapo

N = int(input())        #número de canos

altura_cano = []        #esse comando cria uma lista vazia ----> colchetes = lista

for i in range(N):              #esse for serve pra inputar as alturas dos canos
  altura_cano.append(int(input()))       #a função .append() adiciona um elemento no fim da lista

#len(_): função usada para sabero tamanho de uma lista, ou seja, o número de elementos dela
#[i]: elemento de uma lista, qnd i = 3, [i] = quarto elemento da lista
#o i começa em 0 ---> i = 0, i = 1...

for i in range(len(altura_cano)):                   #esse for serve pra printar as msgs, e acontece no mesmo tempo que o outro
  diff = abs(altura_cano[i+1] - altura_cano[i])     #pra poder fazer esse cálculo, tem que criar um outro for que tenha como range o tamanho da lista
  if (diff > P):
    print("GAME OVER")         #se der mensagem de GAME OVER/YOU WIN, tem que interromper (usando break)
    break                      #e tmb tem que colocar break pra printar msg de GAME OVER ou YOU WIN uma *única* vez
  else:
    print("YOU WIN")
    break
```

```python
#Forma 2

MESSAGE_WON = "YOU WIN"
MESSAGE_LOST = "GAME OVER"

P = int(input())        #corresponde à altura do pulo do sapo
while (P < 0):
  print('O pulo do sapo deve ter uma altura maior que zero')
  P = int(input())

N = int(input())        #corresponde ao número de canos
while (N < 0):
  print('O numero de canos deve ser maior que zero')
  N = int(input())

altura_canos = []

if (N == 1):
  print(MESSAGE_WON)
else:
  for i in range(N):
    altura_cano = int(input())                                    #2ª opção: altura_canos.append(int(input()))
    while (altura_cano < 0):
      print('A altura do cano deve ser maior do que zero')
      altura_cano = int(input())                               #2ª opção: altura_canos.append(int(input()))
    altura_canos.append(altura_cano)                           #2ª opção: nada

  won = True
  for i in range(len(altura_canos)):
    if (i == len(altura_canos)-1):
      continue

    cano_a, cano_b = altura_canos[i], altura_canos[i+1]
    if (abs(cano_b - cano_a) > P):
      won = False
      break

  print( MESSAGE_WON if won else MESSAGE_LOST )
```
