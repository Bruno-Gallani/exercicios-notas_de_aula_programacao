```python
def comparar_datas(d1, m1, a1, d2, m2, a2):
  if a1 == a2:
    if m1 == m2:
      if d1 > d2:
        valor = 1
      elif d1 < d2:
        valor = -1
      elif d1 == d2:
        valor = 0

    elif m1 > m2:
      valor = 1

    elif m1 < m2:
      valor = -1

  elif a1 > a2:
    valor = 1

  elif a2 > a1:
    valor = -1

  return valor
```

##EP 10 - Autonomia do avião

###image.png

```python
import math

carga_do_aviao = int(input())
Ax = float(input())
Ay = float(input())

Bx = float(input())
By = float(input())

def distancia_euclidiana(Ax, Ay, Bx, By):
  distancia_eixoX = (Bx - Ax)**2
  distancia_eixoY = (By - Ay)**2
  distancia_entre_pontos = math.sqrt(distancia_eixoX + distancia_eixoY)
  return distancia_entre_pontos
print("%.2f" % (distancia_euclidiana(Ax, Ay, Bx, By)))

if 0 <= carga_do_aviao <= 50000:
  if distancia_euclidiana(Ax, Ay, Bx, By) <= 18000:
    print("SIM")
  elif distancia_euclidiana(Ax, Ay, Bx, By) > 18000:
    if distancia_euclidiana(Ax, Ay, Bx, By) <= (18000 + (18000*0.1)):
      print("TALVEZ")
    else:
      print("NAO")

elif 50001 <= carga_do_aviao <= 200000:
  if distancia_euclidiana(Ax, Ay, Bx, By) <= 9000:
    print("SIM")
  elif distancia_euclidiana(Ax, Ay, Bx, By) >9000:
    if distancia_euclidiana(Ax, Ay, Bx, By) <= (9000 + (9000*0.1)):
      print("TALVEZ")
    else:
      print("NAO")

elif 200001 <= carga_do_aviao <= 250000:
  if distancia_euclidiana(Ax, Ay, Bx, By) <= 3000:
    print("SIM")
  elif distancia_euclidiana(Ax, Ay, Bx, By) > 3000:
    if distancia_euclidiana(Ax, Ay, Bx, By) <= (3000 + (3000*0.1)):
      print("TALVEZ")
    else:
      print("NAO")
```

##EP 7 - N primeiros primos

###image.png

```python
def eh_primo(x):
  primo = True
  cont = 2
  while x > 2:
    for cont in range(2, x):
      if x % cont == 0:
        return not primo
    return primo       #se passar pelo for, sem ativar a condição, o número é primo (True)
  if x == 2:
    return primo

def main():
  n = int(input())
  numeros_primos = []
  i = 2
  while len(numeros_primos) < n:        #condição de parada: a quantidade de números primos SEMPRE será igual a n
    if eh_primo(i):
      numeros_primos.append(i)
    i += 1
  for numero in numeros_primos:
    print(numero)

main()
```

##EP 2 - Duplas

###image.png

```python
n = int(input("Digite o tamanho do vetor: "))
vetor = []

for i in range(n):
    elemento = int(input("Digite o elemento {}: ".format(i + 1)))
    vetor.append(elemento)

eh_vetor_de_duplas = True
i = 0

while i < n:
    if i == n - 1 or vetor[i] != vetor[i + 1]:         #se o último elemento estiver sem dupla, ou se algum dos elementos consecutivos for diferente entre si
        eh_vetor_de_duplas = False
        break
    else:      #se todos os elementos consecutivos forem iguais
      eh_vetor_de_duplas
    i += 2

if eh_vetor_de_duplas:
    print("É um vetor de duplas!")
else:
    print("Não é um vetor de duplas!")
```
